### Loading Custom Model which is trained on colab by using GPU

```python
model = YOLO(r"custom_model.pt")
```

```python
# Classes on which model has been trained
classes = {0: 'Excavator', 1: 'Gloves', 2: 'Hardhat', 3: 'Ladder', 4: 'Mask', 5: 'NO-Hardhat', 
           6: 'NO-Mask', 7: 'NO-Safety Vest', 8: 'Person', 9: 'SUV', 10: 'Safety Cone', 
           11: 'Safety Vest', 12: 'bus', 13: 'dump truck', 14: 'fire hydrant', 15: 'machinery', 
           16: 'mini-van', 17: 'sedan', 18: 'semi', 19: 'trailer', 20: 'truck and trailer',
           21: 'truck', 22: 'van', 23: 'vehicle', 24: 'wheel loader'}
```

### Using Camera

```python
cap = cv.VideoCapture(0) 
cap.set(3,1500) 
cap.set(4,1000) 
while True: 
    ret, img = cap.read()
    results = model(img,stream=True)
    for result in results:
        boxe = result.boxes
        classes = result.names
        for box in boxe:
            x1,y1,x2,y2 = box.xyxy[0]
            x1,y1,x2,y2= int(x1),int(y1),int(x2),int(y2)
            w,h = x2-x1,y2-y1
            conf = box.conf[0]
            name = classes[int(box.cls[0])]
            if conf >0.4:
                cvzone.cornerRect(img,bbox=[x1,y1,w,h],t=2)
                cvzone.putTextRect(img,f"{name} {math.ceil(conf*100)} %",(max(0,x1),max(35,y1)),scale=1.3,thickness=1,
                               colorR=(120,81,169), font=1,)
            
    cv.imshow('Image', img)
    key = cv.waitKey(1)
    if key == ord('q'):  
        break
cap.release()
cv.destroyAllWindows()
```
